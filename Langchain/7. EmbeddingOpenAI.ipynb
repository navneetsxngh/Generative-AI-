## Enbeddings

- Convert Text Into Vectors

````python
import os
from dotenv import load_dotenv
load_dotenv()
````

````python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
````

````python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small", dimensions=1024)
embeddings
````

````python
text = " This is tutorial on OpenAi embeddings"
query_result = embeddings.embed_query(text)
````

````python
len(query_result)
````

### Using ```speech.txt``` file

````python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("Data/speech.txt")
docs = loader.load()
````

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text = text_splitter.split_documents(docs)
text
````

## Vector Embedding and Vector Store

````python
from langchain_community.vectorstores import Chroma
## Syntax --> Chroma(document, embedding_technique)
db = Chroma.from_documents(text, embeddings)
db
````

````python
query = "t will be all the easier for us to conduct ourselves as belligerents in a high spirit of right and fairness because we act without animus"
retrieved_results = db.similarity_search(query)
retrieved_results
````
